# comparision between Linear Regression and random forest


```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

lr = LinearRegression()
rf = RandomForestRegressor()
```

```python

rf.fit(x_train,y_train)
```

```python
lr.fit(x_train,y_train)
```

```python
y1_pred = lr.predict(x_test)
y2_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error,root_mean_squared_error,r2_score
```

```python
print("errors of Linear Regression")
print("MAE-->",mean_absolute_error(y_test,y1_pred))
print("MSE-->",mean_squared_error(y_test,y1_pred))
print("RMSE-->",root_mean_squared_error(y_test,y1_pred))
print("r2_score-->",r2_score(y_test,y1_pred))
```

```python
print("errors of random forest regressor")
print("MAE-->",mean_absolute_error(y_test,y2_pred))
print("MSE-->",mean_squared_error(y_test,y2_pred))
print("RMSE-->",root_mean_squared_error(y_test,y2_pred))
print("r2_score-->",r2_score(y_test,y2_pred))
```

### Linear regression perform better than the random forest regressor so we should choose the linear regression to train our mode

```python
def generate_recommendations(hours_studied,
                             sleep_hours,
                             attendance_percent,
                             previous_scores):

    recommendations = []

    if hours_studied < 3.5:
        recommendations.append(
            f"Your studied hour is {hour_studied} so Increase study hours to improve academic performance. which will make a positive impact on your marks"
        )

    if sleep_hours < 6:
        recommendations.append(
            "Maintain at least 6-8 hours of sleep daily for better focus."
        )

    if attendance_percent < 75:
        recommendations.append(
            "Improve class attendance and participate regularly other wise your will get the notice."
        )

    if previous_scores < 68:
        recommendations.append(
            "Revise previous concepts and practice more questions."
        )

    return recommendations
```

```python
def analyse_student(hours_studied,sleep_hours,attendance_percent,previous_scores):
    prediction = lr.predict([[hours_studied,sleep_hours,attendance_percent,previous_scores]])[0]
    recommendation = generate_recommendations(hours_studied,
                             sleep_hours,
                             attendance_percent,
                             previous_scores)

    return recommendation,prediction
    
```
